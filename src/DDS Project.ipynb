```python
import os
import sys
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

DATA_PATH = "../data/incarceration_trends_dataset.csv"
```

```python
inc = pd.read_csv(DATA_PATH)
ca_inc = inc[(inc['state'] == 'CA') & (inc['year'] > 1985)].reset_index()
filtered = ca_inc.drop(columns = ['yfips', 'fips', 'commuting_zone', 'metro_area', 'land_area', 'index'])
filtered
```

```python
prison_jail_pop_rates = filtered[['year', 'county_name', 'white_prison_pop_rate', 'native_prison_pop_rate', 'latinx_prison_pop_rate', 'black_prison_pop_rate', 'aapi_prison_pop_rate', 'total_prison_pop_rate', 'white_jail_pop_rate', 'native_jail_pop_rate', 'latinx_jail_pop_rate', 'black_jail_pop_rate', 'aapi_jail_pop_rate', 'total_jail_pop_rate']]
```

```python
jail_pop = filtered[['year', 'county_name', 'total_jail_pop', 'aapi_jail_pop', 'black_jail_pop', 'latinx_jail_pop', 'native_jail_pop', 'white_jail_pop', 'other_race_jail_pop']]
```

```python
dicts = {}

for col in jail_pop.columns:
    if 'pop' in col:
        dicts[col] = jail_pop.fillna(0).groupby('year')[col].sum()
 
jail_pop_races = pd.DataFrame(dicts)
jail_pop_races
```

```python
prison_pop = filtered[['year', 'county_name', 'total_prison_pop', 'aapi_prison_pop', 'black_prison_pop', 'latinx_prison_pop', 'native_prison_pop', 'white_prison_pop', 'other_race_prison_pop']]
```

```python
dictsv2 = {}

for col in prison_pop.columns:
    if 'pop' in col:
        dictsv2[col] = prison_pop.fillna(0).groupby('year')[col].sum()
 
prison_pop_races = pd.DataFrame(dictsv2)
prison_pop_races
```
